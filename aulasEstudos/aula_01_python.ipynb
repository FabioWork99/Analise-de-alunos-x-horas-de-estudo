## 3. Variáveis

Variáveis são espaços usados para guardar valores.

Podemos guardar:
- números
- textos
- resultados de contas

Pense na variável como uma caixa com nome.


````python
# Criando variáveis
nome = "Ana"
idade = 18
temperatura = 27.5

# Mostrando os valores guardados
print(nome)
print(idade)
print(temperatura)
````

### Regra importante
Em Python:
- `=` significa **atribuição**

## 4. Mostrando texto junto com variáveis

Muitas vezes queremos exibir uma frase com os valores das variáveis.


````python
nome = "Carlos"
idade = 20

print("Nome:", nome)
print("Idade:", idade)
print(f"O aluno {nome} tem {idade} anos.")
````

## 5. Operadores de comparação

Antes de aprender estruturas condicionais, precisamos entender comparações.

As comparações geram:
- `True` → verdadeiro
- `False` → falso


````python
print(10 > 5)
print(3 < 1)
print(7 == 7)
print(8 != 2)
print(10 >= 10)
print(4 <= 9)
````

### Operadores usados
- `>` maior que
- `<` menor que
- `==` igual a
- `!=` diferente de
- `>=` maior ou igual
- `<=` menor ou igual

### Observação importante
- `=` atribui valor
- `==` compara valores


## 6. Estrutura condicional `if`

A estrutura `if` serve para tomar decisão.

### Ideia
**Se** uma condição for verdadeira, o Python executa o bloco de código.


````python
temperatura = 32

if temperatura > 30:
    print("Está calor.")
````

### Explicação detalhada
- `if` significa **se**
- `temperatura > 30` é a condição
- se a condição for verdadeira, o código recuado será executado

### Conceito importante: indentação
No Python, os espaços no começo da linha são muito importantes.
Eles mostram qual código pertence ao `if`.


## 7. Estrutura `if` + `else`

O `else` significa **senão**.
Ele é usado quando queremos executar um bloco para o caso contrário.


````python
idade = 16

if idade >= 18:
    print("Maior de idade.")
else:
    print("Menor de idade.")
````

### Como esse código funciona?
1. Python verifica se `idade >= 18`
2. se for verdadeiro, mostra "Maior de idade."
3. se for falso, entra no `else`

### Quando usar?
Quando existem apenas duas possibilidades.


## 8. Estrutura `if` + `elif` + `else`

O `elif` significa **senão se**.
Ele é usado quando temos mais de duas possibilidades.


````python
nota = 7

if nota >= 9:
    print("Excelente")
elif nota >= 7:
    print("Bom")
elif nota >= 5:
    print("Regular")
else:
    print("Precisa melhorar")
````

### Explicação
O Python testa as condições na ordem:
1. nota maior ou igual a 9?
2. se não, nota maior ou igual a 7?
3. se não, nota maior ou igual a 5?
4. se nenhuma condição for verdadeira, entra no `else`

### Dica didática
A ordem das condições importa muito.


## 9. Exemplo com contexto industrial

Vamos aplicar a lógica condicional a uma situação ligada ao curso.

Imagine que estamos monitorando a temperatura de uma máquina.


````python
temperatura_maquina = 85

if temperatura_maquina >= 90:
    print("Alerta crítico: temperatura muito alta.")
elif temperatura_maquina >= 75:
    print("Atenção: temperatura acima do ideal.")
else:
    print("Temperatura dentro do esperado.")
````

### Relação com a área de dados
Esse tipo de código pode ser usado para:
- gerar alertas
- classificar medições
- tomar decisões automáticas
- preparar regras simples antes de análises mais avançadas

## 12. Estruturas de repetição

As estruturas de repetição servem para executar uma ação várias vezes.

Nesta aula, vamos trabalhar com:
- `while`
- `for`


## 13. Repetição com `while`

O `while` significa:
**enquanto uma condição for verdadeira, repita o bloco**


````python
contador = 1

while contador <= 5:
    print(contador)
    contador = contador + 1
````

### Explicação linha a linha
- `contador = 1` cria a variável de controle
- `while contador <= 5:` diz que o laço continua enquanto a condição for verdadeira
- `print(contador)` mostra o valor atual
- `contador = contador + 1` aumenta o contador em 1

### Atenção
Se esquecermos de atualizar o contador, podemos criar um laço infinito.


## 14. Exemplo de `while` com contexto simples

Vamos simular o registro de peças produzidas.


````python
peca = 1

while peca <= 5:
    print(f"Peça {peca} registrada.")
    peca = peca + 1
````

## 15. Repetição com `for`

O `for` é usado para percorrer uma sequência.

Ele é muito comum quando queremos:
- repetir uma ação um certo número de vezes
- percorrer listas
- analisar vários valores


````python
for numero in range(5):
    print(numero)
````

### Explicação
- `range(5)` gera os números: 0, 1, 2, 3, 4
- o `for` percorre esses valores um por um
- a variável `numero` recebe cada valor da sequência


````python
for numero in range(1, 6):
    print(numero)
````

### Diferença
- `range(5)` começa no 0
- `range(1, 6)` começa no 1 e vai até o 5


## 16. `for` com lista

Também podemos usar `for` para percorrer listas.


````python
maquinas = ["M1", "M2", "M3"]

for maquina in maquinas:
    print(f"Analisando a máquina {maquina}")
````

### Explicação
A variável `maquina` recebe, uma por vez:
- M1
- M2
- M3

Isso é muito útil quando queremos analisar vários itens.


## 17. Combinando repetição e condição

Na prática, muitas vezes usamos repetição e decisão juntas.


````python
notas = [6, 8, 4, 9, 7]

for nota in notas:
    if nota >= 7:
        print(f"Nota {nota}: aprovado")
    else:
        print(f"Nota {nota}: estudar mais")
````

### O que aconteceu aqui?
- o `for` percorreu cada nota da lista
- o `if` decidiu o que mostrar para cada valor

Essa combinação é muito importante para análise de dados.


## 18. Exemplo com contexto industrial

Agora vamos simular temperaturas de máquinas.


````python
temperaturas = [68, 72, 91, 77, 95]

for temp in temperaturas:
    if temp >= 90:
        print(f"{temp}°C -> Alerta crítico")
    elif temp >= 75:
        print(f"{temp}°C -> Atenção")
    else:
        print(f"{temp}°C -> Normal")
````

### Relação com a área de dados
Esse tipo de lógica pode ser usado para:
- percorrer medições
- gerar classificações simples
- criar alertas iniciais
- preparar dados para relatórios
