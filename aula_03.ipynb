```python
#Fazendo a importação da biblioteca necessária e dando um alias para facilitar o desenvolvimento.
import pandas as pd
```

```python
#Por questão de boas práticas, a variável responsável por receber o dataset se chamará df.
df = pd.read_csv(r".\data\dataset_aula1_numpy_pandas.csv")
```

```python
#Ao chamar a variável que recebeu o dataset, os dados serão exibidos de forma formatada.
df
```

```python
#Por padrão a função head exibe as primeiras 5 linhas do dataset. Ao inserir um número dentro do parênteses, o número inserido será o número de linhas.
df.head()
```

```python
#O tail vai exibir as últimas linhas do dataset, por padrão ele exibe as 5 últimas.
df.tail()
```

```python
#O shape é uma ferramenta que vai exibir as dimensões do dataset, sendo elas: (quantidade de linhas no dataset, quantidade de colunas no dataset).
df.shape
```

```python
#info() é uma função que exibe as informações sobre o dataset.
df.info()
```

```python
#columns é uma ferramenta que exibe as colunas presentes no dataset.
df.columns
```

```python
#o .isna() vai procurar por valores que estão nulos e o .sum vai fazer a soma de todos esses valores nulos.
df.isna().sum()
```

```python
#Exibe os valores não nulos.
df.notnull()
```

```python
#Exibe só os valoes de uma coluna específica.
df["temperatura_c"]
```

```python
#Para exibir mais de uma coluna, é preciso fazer dentro de mais um colchete.
df[["temperatura_c", "setor"]]
```

```python
#Comando para visualizar as colunas em formato de lista.
df.columns.to_list()
```

```python
df.describe()
```
