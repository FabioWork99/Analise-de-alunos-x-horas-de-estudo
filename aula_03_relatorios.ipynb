## 2) Novas analises

### A qualidade da dieta do aluno (diet_quality) faz diferença na nota? Mostre a média por grupo.

Olhando puramente para os números, a qualidade da dieta não faz uma diferença expressiva na nota final do exame.

Embora os alunos com uma dieta ruim (Poor) tenham uma média ligeiramente menor, a variação entre os três grupos é muito sutil (menos de 2,5 pontos de diferença entre a maior e a menor média). Além disso, o desvio padrão em todos os grupos é de aproximadamente 17 pontos, o que significa que há uma alta variação de notas dentro de cada próprio grupo.

#### Alunos com trabalho de meio período (part_time_job) vão pior nas notas?

Não, os alunos que trabalham em meio período não vão significativamente pior nas notas.

A diferença entre as notas de quem trabalha e quem não trabalha é praticamente desprezível (pouco mais de 1 ponto na média).
Ambos os grupos apresentam um comportamento muito parecido, com a mesma dispersão de notas (desvio padrão em torno de 16 pontos), indicando que o fato de ter um emprego de meio período não está prejudicando o desempenho final no exame de forma relevante.

### A escolaridade dos pais (parental_education_level) tem relação com o desempenho?

Assim como a dieta e o trabalho de meio período, a escolaridade dos pais também não apresenta uma relação forte ou relevante com o desempenho dos alunos no exame.

Os dados revelam que as médias de notas entre os diferentes níveis de escolaridade dos pais são extremamente próximas, variando em uma faixa de apenas 2 pontos. Curiosamente, os alunos cujos pais possuem mestrado (Master) tiveram uma média ligeiramente menor do que aqueles cujos pais possuem apenas o ensino médio (High School), mas essa diferença é tão pequena dentro da variação geral que é considerada estatisticamente insignificante.

## 3. Cuidados e Ética no uso da IA

Tão importante quanto coletar dados e informação da IA é conferir e analisar, muitas vezes a inteligencia artificial se engana e confunde, cabe ao analista verificar e analisar as infomações cedidas ao usuario.

Existem perigos em enviar um trabalho feitos por IA, comprometimento com as veracidades presentes no trabalho é respeito ao leitor e clientes, atinge diretamente na credibilidade do analista.

Por que é perigoso confiar cegamente na IA no trabalho?
A responsabilidade final é sempre humana: Como aponta o texto, a inteligência artificial pode se enganar, confundir contextos ou simplesmente assumir premissas falsas como verdadeiras. Cabe ao analista atuar como um filtro crítico.



Risco de Alucinação: Modelos de linguagem são projetados para gerar textos convincentes. Se um profissional aceita um relatório gerado por IA sem validar as fontes ou os dados brutos, ele corre o risco de repassar informações completamente inventadas para diretores ou clientes.



Impacto na credibilidade: Entregar um trabalho com dados errados compromete diretamente a reputação técnica e a ética do analista perante o mercado e seus clientes. O respeito ao leitor exige precisão.
